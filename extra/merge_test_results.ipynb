```python
tasks = {
    "bin": {
        "exp": "lr-2e-05-optimizer-Adafactor-epoch-10"
    },
    "multi-label": {
        "exp": "lr-2e-05-optimizer-AdamW-epochs-10-RMSEduringTraining"
    },
    "reg": {
        "exp": "lr-2e-05-optimizer-AdamW-epochs-10-RMSEduringTraining"
    }
}
```

```python
import os
import pandas as pd
PATH = "/content/drive/My Drive/HUHU-IberLEF2023/"

for task, content in tasks.items():
  current_path = os.path.join(PATH, task, "outputs", content.get("exp"))
  tasks[task]["test"] = pd.read_csv(os.path.join(current_path, "test", "results.csv"))
```

```python
predictions = pd.merge(tasks["bin"]["test"], tasks["multi-label"]["test"], on="tweet_id")
predictions = pd.merge(predictions, tasks["reg"]["test"], on="tweet_id")

print("Predictions dataset size:", len(predictions.index))
predictions.head()
```

```python
# Save results in CSV file
predictions.to_csv(os.path.join(PATH, "HUHUNLP_Submission2.csv"), index=False)
```
