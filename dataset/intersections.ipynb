```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dict_label = {"prejuicio-mujeres": {"prejuicio-mujeres": 822, "prejuicio-lgbtiq": 14, "prejuicio-inmigrantes": 2, "gordofobia": 58},
              "prejuicio-lgbtiq": {"prejuicio-lgbtiq": 415, "prejuicio-mujeres": 14, "prejuicio-inmigrantes": 1, "gordofobia": 6},
              "prejuicio-inmigrantes": {"prejuicio-inmigrantes": 459, "prejuicio-mujeres": 2, "prejuicio-lgbtiq": 1, "gordofobia": 2},
              "gordofobia": {"gordofobia": 90, "prejuicio-mujeres": 58, "prejuicio-lgbtiq": 6, "prejuicio-inmigrantes": 2}}
```

```python
df = pd.DataFrame(dict_label.values())
df.head()
```

```python
df_norm = df/df.sum().sum()
df_norm.head()
```

```python
LABELS = dict_label.keys()
fig_cm = plt.figure(figsize = (10.8, 10.8))
mask = np.triu(np.ones_like(df, dtype = bool))
np.fill_diagonal(mask, False)

sns.heatmap(df,
            vmin = df.min().min(),
            vmax = df.max().max(),
            square = True,
            annot = True, fmt = ".0f",
            cbar_kws = {"shrink" : 0.5},
            xticklabels = LABELS,
            yticklabels = LABELS,
            annot_kws = {"size": 15},
            mask = mask)

plt.xlabel("", fontsize = 15)
plt.ylabel("", fontsize = 15)
plt.title("Distribucin de intersecciones", fontsize = 20, pad = 10)
plt.show()
```
